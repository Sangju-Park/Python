```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('rhj4.jpg')
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
expand=cv2.resize(image,dsize=None,fx=3.0,fy=3.0,interpolation=cv2.INTER_CUBIC)
plt.imshow(cv2.cvtColor(expand,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('rhj7.jpg')

# 행과 열 정보만 저장합니다.
height, width = image.shape[:2]

M = np.float32([[1, 0, 50], [0, 1, 10]])
dst = cv2.warpAffine(image, M, (500, 400))

plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
plt.show()
```
