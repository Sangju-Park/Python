```python
#본 Page
url="https://search.naver.com/search.naver?where=image&sm=tab_jum&query=유현주+프로"
res=req.get(url)
soup=bs(res.content,'html.parser')
```

```python
#하위 Url
data=''.join(str(soup))
p=r'https?://[\.\-&/?=\w]*\d{3,}'
lists=re.findall(p,data)
```

```python
#하위Url 필터링
snd=[]
for list in lists:
    res=req.get(list)
    if res.status_code==200:
        soup=bs(res.content,'html.parser')
        tem=soup.select('img')
        snd.append(tem)
```

```python
data=''.join(str(snd))

p=r'https?://[\.\-&/?=\w]*\d{3,}'
lists=re.findall(p,data)
```
